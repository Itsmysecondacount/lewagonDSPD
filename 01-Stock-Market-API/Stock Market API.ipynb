```python
stocks = ['goog', 'fb', 'amzn', 'aapl', 'meli']
stocks_data = pd.DataFrame()

for stock in stocks:
    url = f"http://iex.lewagon.com/stable/stock/{stock}/chart/1y"
    api_data = pd.DataFrame(requests.get(url).json())
    
    #fb NO trae data por eso existía el error
    if not api_data.empty:
        api_data['date'] = pd.to_datetime(api_data['date'])
        api_data.set_index('date', inplace=True)
        # Solo selecciona la columna 'close' y la renombra al nombre del stock
        stocks_data[stock] = api_data['close']
    
stocks_data.head()
```

```python
stocks_data.plot(figsize=(16,6))
```

```python
#normalization technique
def normalization(series):
    return series/series[0]

stocks_data_normalized = stocks_data.apply(normalization, axis=0)

print(stocks_data_normalized.head())
stocks_data_normalized.plot(figsize=(15,7))
```

```python
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib
```

```python
url = "http://books.toscrape.com"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
soup
```

Inspect and scrape!

```python
books_html = soup.find_all("article", class_="product_pod")


# print(books_html)
len(books_html)
```

```python
books_html
```

## Parsing _one_ book

```python
books_html[0]
```

```python
book_title = books_html[0].find("h3").find("a")
book_title
```

```python
book_title.attrs
```

```python
book_title = books_html[0].find("h3").find("a")['title']
book_title
```

```python
book_price = float(books_html[0].find("p", class_="price_color").string.strip("£"))
book_price
```

Examine how stars are tagged:

```python
book_stars_html = books_html[0].find("p", class_="star-rating").attrs
book_stars_html
```

```python
book_stars = book_stars_html['class'][1]
book_stars
```

```python
def parse_rating(rating_classes):
    ratings = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    if rating_classes[1] in ratings.keys():
        return ratings[rating_classes[1]]
    else:
        return 0
```

```python
parse_rating(book_stars_html['class'])
```

## Parsing _all_ books

```python
books_dict = {'Title': [], 'Price': [], 'Rating': []}
```

```python
for book in books_html:
    books_dict['Title'].append(book.find("h3").find("a")['title'])
    books_dict['Price'].append(float(book.find("p", class_="price_color").string.strip("£")))
    books_dict['Rating'].append(parse_rating(book.find("p", class_="star-rating").attrs['class']))

books_dict
```

```python
len(books_dict)
```

```python
len(books_dict['Title'])
```

## Loading data into _Pandas_

```python
books_df = pd.DataFrame.from_dict(books_dict)
books_df
```

```python
books_df.groupby("Rating").count()
```

```python
books_df.groupby("Rating").count()['Title'].plot(kind='bar')
```

### Going through _all_ the pages of the catalogue

```python
MAX_PAGE = 20
all_books_dict = {'Title': [], 'Price': [], 'Rating': []}

for page in range(1,MAX_PAGE + 1):
    print(f"Parsing page {page}/{MAX_PAGE}...")
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    for book in soup.find_all("article", class_="product_pod"):
        all_books_dict['Title'].append(book.find("h3").find("a")['title'])
        all_books_dict['Price'].append(float(book.find("p", class_="price_color").string.strip("£")))
        all_books_dict['Rating'].append(parse_rating(book.find("p", class_="star-rating").attrs['class']))

all_books_dict
```

```python
len(all_books_dict["Title"])
```

```python
all_books_df = pd.DataFrame.from_dict(all_books_dict)
all_books_df.tail()
```

```python
all_books_df["Price"].hist()
```

```python
all_books_df.groupby("Rating").count()["Title"].plot(kind="bar")
```

#### Save scraping results!!
